<br>What is a recommender system:
   * Based on previous(past) user-item interactions (behaviors), it predicts the likelihood that a user would prefer an item.
   * For example, movie recommendation system suggests people new movies according to their past activities like watching and voting movies.
   * The purpose of recommender system is recommending new things that are not seen before from people (interest exploration).
   
<br>The most classical recommender system:
1. User Based Collaborative Filtering
    * Collaborative filtering is making recommend according to combination of your experience and experiences of other people.
    * First we need to make user vs item matrix.
        * Each row is users and each columns are items like movie, product or websites
    * Secondly, computes similarity scores between users.
        * Each row is users and each row is vector.
        * Compute similarity of these rows (users).
    * Thirdly, find users who are similar to you based on past behaviours
    * Finally, it recommends what you have not experienced before.
    * An example of user-based collaborative filtering:
        * Think that there are two people
        * First people watched 2 movies: lord of the rings and hobbit
        * Second people watched only lord of the rings movie
        * User based collaborative filtering computes similarity of these two people and sees both of them have watched the movie lord of the rings.
        * So the model identified the similarity between these two people. So it will recommend the hobbit movie to second people as it can be seen from the picture below:
        *<a href="https://ibb.co/droZMy"><img src="https://preview.ibb.co/feq3EJ/resim_a.jpg" alt="resim_a" border="0"></a>
        
    * However, User-based collaborative filtering does have some problems and drawbacks:
        * In this system, each row of matrix represents a user. Therefore, computing, comparing, and finding similarity between of them is computationaly expensive and will spend too much computation.
        * Also, habits of people can be changed. Therefore making correct and useful recommendation can be hard in time.
    * In order to solve these problems, the item-based collaborative filtering approach was proposed.

1. Item-based Collaborative Filtering
    * In this method, instead of finding relationship between users, user-interacted items like movies or stuffs are compared with each other to find item similarities.
    * In user-based recommendation system, habits of users can be changed. This situation makes the system hard to adapt. However, in the item-based recommendation system, movies or stuffs does not change quickly over time. Therefore this recommendation performance is relatively stable.
    * There are almost 7 billion people (possible users) all over the world. Comparing people similarities (in most cases) is much harder in computation cost than computing item similarities.
    * In thee item-based recommendation system, we still need to make user vs item matrix that is also used in user-based recommender system.
        * Each row represents a user and each column represents an item like movie, product or website.
        * However, at this time, instead of calculating similarity between rows, the model needs to calculate similarity between columns that are items like movies or stuffs.
    * An example of how it works:
        * Firstly, there is similarity between the movie lord of the rings and hobbit because both are liked by three different people. There is a similarity point between these two movies.
        * If the similarity is high enough, we can recommend hobbit to other people who only watched lord of the rings as it can be seen in figure below.
        *<a href="https://imgbb.com/"><img src="https://image.ibb.co/maEQdd/resim_b.jpg" alt="resim_b" border="0"></a>




```python
import pandas as pd

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

import os
print(os.listdir("../input/movielens-20m-dataset"))
```

```python
# import movie data set and look at columns
movie = pd.read_csv("../input/movielens-20m-dataset/movie.csv")
movie.columns
```

```python
# what we need are movie id and title columns
movie = movie.loc[:,["movieId","title"]]
movie.head(10)
```

```python
# import rating data and look at columns
rating = pd.read_csv("../input/movielens-20m-dataset/rating.csv")
rating.columns
```

```python
# what we need are user id, movie id and rating columns
rating = rating.loc[:,["userId","movieId","rating"]]
rating.head(10)
```

```python
# then merge movie and rating data
data = pd.merge(movie,rating)
```

```python
# now lets look at our data 
data.head(10)
```

* As it can be seen data frame that is above, we have 4 features that are movie id, title user id and rating
* According to these data frame, we will make item based recommendation system
* Lets look at shape of the data. The number of sample in data frame is 20 million that is too much. There can be problem in kaggle even if their own desktop ide's like spyder or pycharm.
* Therefore, in order to learn item based recommendation system lets use 1 million of sample in data.

```python
data.shape
```

```python
data = data.iloc[:1000000,:]
```

```python
data.shape
data.head()
```

```python
# Lets make a pivot table in order to make rows are users and columns are movies, the value in this table are rating given by the user to this certain movie
pivot_table = data.pivot_table(index = ["userId"],columns = ["title"],values = "rating")
pivot_table.head(10)
```

As it can be seen from table above, rows are users, columns are movies and values are ratings. 

For example, userid 11 gives 3.5 rating to the movie "Ace Ventura: When Nature Calls (1995)" and gives a 3.0 rating to movie "Bad Boys (1995)".

* Now let's resolve a scenario:
    * On the movie web site, "Bad Boys (1995)" movie are watched and rated by people. The question is that which movie should the recommender system recommend (push) to audience who have watched "Bad Boys (1995)" movie?
    
* In order to answer this question, we need to find similarities between "Bad Boys (1995)" movie and other movies.

```python
movie_watched = pivot_table["Bad Boys (1995)"]
movie_watched
```

```python
movie_watched = pivot_table["Bad Boys (1995)"]
# find correlation between "Bad Boys (1995)" and other movies
similarity_with_other_movies = pivot_table.corrwith(movie_watched)
# sort the similarities from the highest to the lowest
similarity_with_other_movies = similarity_with_other_movies.sort_values(ascending=False)
# output the top 5 results
similarity_with_other_movies.head(6)
```

* It can be concluded that we need to recommend "Headless Body in Topless Bar (1995)" movie to people who watched "Bad Boys (1995)".
* On the other hand, the number of rating for each movie is also important.

```python
movie_watched = pivot_table["Two Bits (1995)"]
# find correlation between "Two Bits (1995)" and other movies
similarity_with_other_movies = pivot_table.corrwith(movie_watched)
# sort the similarities from the highest to the lowest
similarity_with_other_movies = similarity_with_other_movies.sort_values(ascending=False)
# output the top 6 results
similarity_with_other_movies.head(6)
```
